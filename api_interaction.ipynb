```python
import requests
import json
import pandas as pd


base_url = 'http://127.0.0.1:5001'

def get_response(endpoint):
    response = requests.get(f'{base_url}/{endpoint}').json()
    data_response = pd.DataFrame(response)
    return data_response.head(15)

```
